## Q1. How stable are top-of-book prices (best bid/ask)?

## Q2. How tight is the spread over time?

```python
crwv["spread"].plot(color="purple", marker="s")
plt.title("Bid-Ask Spread Over Time")
plt.grid(True)
plt.show()
```

## Q3. How does liquidity (depth) change over time at top levels?

```python
plt.figure(figsize=(10, 5))
# Compute top 5 depth
bid_sz_cols = [f"bid_sz_0{i}" for i in range(5)]
ask_sz_cols = [f"ask_sz_0{i}" for i in range(5)]
crwv["bid_depth_top5"] = crwv[bid_sz_cols].sum(axis=1)
crwv["ask_depth_top5"] = crwv[ask_sz_cols].sum(axis=1)

# Imbalance
crwv["depth_imbalance"] = crwv["bid_depth_top5"] / (crwv["bid_depth_top5"] + crwv["ask_depth_top5"])

# Rolling volatility
#crwv = crwv.sort_values("timestamp").set_index("timestamp")
crwv["mid_price_return"] = crwv["mid_price"].pct_change()
crwv["rolling_volatility"] = crwv["mid_price_return"].rolling(window=3).std()

crwv["bid_depth_top5"].plot(label="Top 5 Bid Depth", marker="^")
crwv["ask_depth_top5"].plot(label="Top 5 Ask Depth", marker="v")
plt.title("Cumulative Depth at Top 5 levels")
plt.legend()
plt.grid(True)
plt.show()
```

Liquidity on both bid and ask sides fluctuates widely. The ask side shows occasional massive spikes in depth (above 35,000), while the bid side sometimes drops near zero. These imbalances suggest that large trades should be timed to match higher depth windows for minimal slippage.

# Modeling g(x)

Q1. What form should g(x) take?
- Linear: gt(x) = Btx (Fails to capture non-linear depth dynamics)
- Piecewise linear or spline based (Captures varying slippage across volume ranges (eg low slippage up to depth 1 then steeper))
- Empirical (non parametric) (No smooth function and hard to generalize or optimize over)
- Power Law: gt(x) = ax^y (almgren-chriss/bouchaud, needs fitting, parameter estimation can be noisy on short windows)

Q2. How do we define slippage precisely?
- Mid-price: Slippage = VWAPbuy - mid (Sensitive to rapid price changes)
- VWAP vs Pre/Post-Trade Mid (Could incorporate market impact beyond instantaneous execution)

Q3. What quantity x range should we model over?
- Relative size: Normalizes across time and instruments 


```python
crwv = pd.read_csv("Data/CRWV/CRWV_2025-04-03 00:00:00+00:00.csv")
crwv["ts_event"] = pd.to_datetime(crwv["ts_event"])

# Sort and set index
crwv = crwv.sort_values("ts_event")
crwv.set_index("ts_event", inplace=True)

# Sample the last LOB update within each 1 min interval
crwv_1min = crwv.groupby(pd.Grouper(freq="1min")).tail(1).reset_index()

# Display sample shape and preview
crwv_1min.shape, crwv_1min[["ts_event", "symbol"]].head()
```

```python
# Select random 10 samples for visualization

sample_snapshots = crwv_1min.sample(n = 10, random_state=42).sort_values("ts_event").reset_index(drop=True)

# Identify column sets
ask_price_cols = [col for col in crwv_1min.columns if col.startswith("ask_px_")]
ask_size_cols = [col for col in crwv_1min.columns if col.startswith("ask_sz_")]
bid_price_cols = [col for col in crwv_1min.columns if col.startswith("bid_px_")]
bid_size_cols = [col for col in crwv_1min.columns if col.startswith("bid_sz_")]

# Prepare storage for all curves
slippage_curves = []

# Absolute sizes for simulation
abs_sizes = np.linspace(1, 2000, 20)
```

```python
# VWAP simluation function

def simulate_vwap(prices, sizes, target_qty):
    vwap = 0
    filled = 0
    for p, s in zip(prices, sizes):
        fill = min(s, target_qty - filled)
        vwap += fill * p
        filled += fill
        if filled >= target_qty:
            break
    if filled < target_qty:
        return np.nan
    return vwap/target_qty

# Simulate slipping for wach sample snapshot
for idx, row in sample_snapshots.iterrows():
    ask_prices = [row[col] for col in ask_price_cols]
    ask_sizes = [row[col] for col in ask_size_cols]
    bid_prices = [row[col] for col in bid_price_cols]
    bid_sizes = [row[col] for col in bid_size_cols]

    mid_price = (row["bid_px_00"] + row["ask_px_00"]) / 2

    buy_slip, sell_slip = [], []
    for x in abs_sizes:
        vwap_buy = simulate_vwap(ask_prices, ask_sizes, x)
        vwap_sell = simulate_vwap(bid_prices, bid_sizes, x)
        g_buy = vwap_buy - mid_price if not np.isnan(vwap_buy) else np.nan
        g_sell = mid_price - vwap_sell if not np.isnan(vwap_sell) else np.nan
        buy_slip.append(g_buy)
        sell_slip.append(g_sell)
    
    slippage_curves.append({
        "timestamp": row["ts_event"],
        "abs_size": abs_sizes,
        "buy_slippage": buy_slip,
        "sell_slippage": sell_slip,
    })

# Plot the curves
fig, axes = plt.subplots(2, 1, figsize=(12, 10), sharex=True)

for curve in slippage_curves:
    axes[0].plot(curve["abs_size"], curve["buy_slippage"], alpha=0.7, label=curve["timestamp"].time())
    axes[1].plot(curve["abs_size"], curve["sell_slippage"], alpha=0.7,label=curve["timestamp"].time())

axes[0].set_title("Buy-side Slippage gᵗᵇ(x) for 10 Random Timestamps")
axes[1].set_title("Sell-side Slippage gᵗˢ(x) for 10 Random Timestamps")
axes[1].set_xlabel("Order Size (shares)")
axes[0].set_ylabel("Buy Slippage ($)")
axes[1].set_ylabel("Sell Slippage ($)")
axes[0].legend(loc="upper left", fontsize="small")
axes[1].legend(loc="upper left", fontsize="small")
for ax in axes: ax.grid(True)
plt.tight_layout()
plt.show()
```

# 📘 Optimization Formulation to Minimize Temporary Market Impact

## 🎯 Objective

Given:
- **S**: Total number of shares to buy (known at start of day)
- **N = 390**: Number of trading periods (1-minute intervals)
- **gₜ(xₜ)**: Temporary impact (slippage) function for trading `xₜ` shares at time `t`

We want to minimize total slippage over the day by choosing an allocation vector:

$$\mathbf{x} = [x_1, x_2, \dots, x_N] \in \mathbb{R}_{\ge 0}^N $$

### Problem:

$$
\min_{\mathbf{x}} \sum_{t=1}^{N} g_t(x_t)
\quad \text{subject to} \quad \sum_{t=1}^{N} x_t = S
$$

---

## 📐 Modeling the Temporary Impact Function

From data, we model each impact function as a power law:


$$ g_t(x_t) = \alpha_t x_t^{\gamma_t} $$

- \( $ \alpha_t > 0 $ \): scales impact at time t  
- \( $ \gamma_t > 1 $ \): controls convexity (e.g., 1.1–1.5 typical)

---

## ✅ Final Optimization Problem

$$
\min_{\{x_t\}_{t=1}^{390}} \sum_{t=1}^{390} \alpha_t x_t^{\gamma_t}
\quad \text{subject to} \quad \sum_{t=1}^{390} x_t = S,\quad x_t \ge 0 \ \forall t
$$

---

## ✏️ Optional Constraints

- **Max Participation Rate**:
  
  $$ x_t \le \eta \cdot D_t$$ 

  where \( $ D_t $ \) is top-10 depth at time t, and \( $ \eta \in [0, 1] $ \)

- **Volatility Penalty**:
  $$
  + \lambda \sum_{t=1}^{390} \sigma_t x_t^2
  $$
  where \( $ \sigma_t $ \) is per-minute mid-price volatility

---

## 🧠 Solving Strategy

- Problem is **convex** if \( $ \gamma_t \ge 1 $ \)
- Use standard solvers: CVXPY, SciPy, or custom Lagrangian method


```python
import cvxpy as cp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Setup
N = 390                    # number of intervals
S = 100000                 # total shares to buy

# Example: replace these with your real estimates
np.random.seed(42)
alpha = np.random.uniform(1e-6, 1e-5, N)
gamma = np.random.uniform(1.1, 1.5, N)

# Variable
x = cp.Variable(N)

# Objective
objective = cp.Minimize(cp.sum([alpha[t] * cp.power(x[t], gamma[t]) for t in range(N)]))

# Constraints
constraints = [cp.sum(x) == S, x >= 0]

# Solve
problem = cp.Problem(objective, constraints)
problem.solve()

# Output
if x.value is not None and np.all(x.value >= 0):
    solution = pd.DataFrame({
        "minute": np.arange(N),
        "x_t": x.value,
        "alpha_t": alpha,
        "gamma_t": gamma
    })
else:
    print("Optimization failed or returned invalid solution.")
    solution = None

# Plot
plt.figure(figsize=(10, 4))
plt.plot(solution["minute"], solution["x_t"], label="Optimal Trade Allocation")
plt.title("Optimal xₜ Allocation Over 390 Minutes")
plt.xlabel("Minute")
plt.ylabel("Shares to Buy")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit

# Load your CRWV file
df = pd.read_csv("Data/CRWV/CRWV_2025-04-03 00:00:00+00:00.csv", parse_dates=["ts_event"])
df = df.sort_values("ts_event")
df.set_index("ts_event", inplace=True)
df_1min = df.groupby(pd.Grouper(freq="1min")).tail(1).reset_index()

# Column helpers
ask_price_cols = [c for c in df.columns if c.startswith("ask_px_")]
ask_size_cols = [c for c in df.columns if c.startswith("ask_sz_")]
bid_price_cols = [c for c in df.columns if c.startswith("bid_px_")]
bid_size_cols = [c for c in df.columns if c.startswith("bid_sz_")]

# VWAP simulator
def simulate_vwap(prices, sizes, qty):
    filled = 0
    vwap = 0
    for p, s in zip(prices, sizes):
        fill = min(qty - filled, s)
        vwap += fill * p
        filled += fill
        if filled >= qty:
            break
    return vwap / qty if filled >= qty else np.nan

# Power law
def power_law(x, alpha, gamma):
    return alpha * (x ** gamma)

# Fit alpha, gamma
results = []
sample = df_1min.sample(n=100, random_state=42)

for _, row in sample.iterrows():
    try:
        apx = [row[c] for c in ask_price_cols]
        asz = [row[c] for c in ask_size_cols]
        mid = (row["bid_px_00"] + row["ask_px_00"]) / 2

        xs = np.linspace(100, 2000, 20)
        ys = []
        for x in xs:
            vwap = simulate_vwap(apx, asz, x)
            ys.append(vwap - mid if not np.isnan(vwap) else np.nan)

        x_fit = xs[~np.isnan(ys)]
        y_fit = np.array(ys)[~np.isnan(ys)]

        if len(x_fit) >= 5:
            popt, _ = curve_fit(power_law, x_fit, y_fit, maxfev=10000)
            results.append({"ts_event": row["ts_event"], "alpha": popt[0], "gamma": popt[1]})
    except:
        continue

pd.DataFrame(results).to_csv("alpha_gamma_CRWV_2025-04-03.csv", index=False)
```
